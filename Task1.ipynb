```python

import requests
import pandas as pd
import matplotlib.pyplot as plt

# Open-Meteo API (No API key required)
url = "https://api.open-meteo.com/v1/forecast?latitude=17.3850&longitude=78.4867&hourly=temperature_2m"

response = requests.get(url)
data = response.json()

time = data['hourly']['time'][:24]
temperature = data['hourly']['temperature_2m'][:24]

df = pd.DataFrame({
    'Time': time,
    'Temperature': temperature
})

print(df.head())

plt.figure(figsize=(10,5))
plt.plot(df['Time'], df['Temperature'], marker='o')
plt.title("24-Hour Temperature Forecast")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
